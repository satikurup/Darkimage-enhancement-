```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('h.jpg')
imgg=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
median = cv2.medianBlur(imgg,5)

plt.subplot(),plt.imshow(imgg),plt.title('Original')
plt.xticks([121]), plt.yticks([])
plt.subplot(),plt.imshow(median),plt.title('Blurred')
plt.xticks([122]), plt.yticks([])
plt.show()
cv2.imwrite('./result.png', median)
```

```python
img = cv2.imread('result.png') 
  
# Trying 4 gamma values. 
for gamma in [ 0.5]: 
      
    # Apply gamma correction. 
    gamma_corrected = np.array(255*(img / 255) ** gamma, dtype = 'uint8') 
  
    plt.subplot(121),plt.imshow(img),plt.title('Original')
    plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(gamma_corrected),plt.title('1st')
    plt.xticks([]), plt.yticks([])
    plt.show()
    cv2.imwrite('m.png', gamma_corrected)
```

```python
from PIL import Image
from PIL import ImageEnhance
def adjust_brightness(input_image, output_image, factor):
    image = Image.open(input_image)
    enhancer_object = ImageEnhance.Brightness(image)
    out = enhancer_object.enhance(factor)
    out.save(output_image)
if __name__ == '__main__':
    adjust_brightness('m.png',
                      'lighthouse_darkened.jpg',1.75)
```

```python
img = cv2.imread('lighthouse_darkened.jpg')
plt.subplot(121),plt.imshow(img),plt.title('1st')
plt.xticks([]), plt.yticks([])
plt.show()
 
```
